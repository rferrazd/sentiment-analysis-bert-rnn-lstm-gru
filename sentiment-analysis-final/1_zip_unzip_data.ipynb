# Notebook to unzip/zip datasets due to github's file size limit

```python
zip_name = 'data'
```

```python
import os
import zipfile

# ZIP
# Create zip directory if it doesn't exist
if not os.path.exists(zip_name):
    os.makedirs(zip_name)

# Copy .pkl files into the zip directory
for filename in ["train.pkl", "val.pkl", "test.pkl", "clean_df.pkl"]:
    if os.path.exists(filename):
        dest_path = os.path.join(zip_name, filename)
        if not os.path.exists(dest_path):
            with open(filename, "rb") as src, open(dest_path, "wb") as dst:
                dst.write(src.read())

# Zip the files in the folder
zipf_path = f"{zip_name}.zip"
with zipfile.ZipFile(zipf_path, "w", zipfile.ZIP_DEFLATED) as zipf:
    for file in ["train.pkl", "val.pkl", "test.pkl"]:
        file_path = os.path.join(zip_name, file)
        if os.path.exists(file_path):
            zipf.write(file_path, arcname=file)

print(f"Zipped files into {zipf_path}")
```

```python
# UNZIP
import pickle

with zipfile.ZipFile(zipf_path, "r") as zipf:
    zipf.extractall(zip_name)
print(f"Extracted files to {zip_name}/")

# (Optional) Load datasets from the unzip folder to confirm
with open(os.path.join(zip_name, "train.pkl"), "rb") as f:
    train = pickle.load(f)
with open(os.path.join(zip_name, "val.pkl"), "rb") as f:
    val = pickle.load(f)
with open(os.path.join(zip_name, "test.pkl"), "rb") as f:
    test = pickle.load(f)
```
